```python
new_meta['id'] = new_meta['id'].astype(int)
```

```python
data = pd.merge(new_meta, credits, on='id')
data.head()
```

```python
pd.set_option('display.max_colwidth', 75)
data
```

```python
# lest change the data as python supports
import ast
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
data.info()
```

```python
data['genres'][0]
```

```python
def make_genresList(x):
    gen = []
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (' '.join(gen))
```

```python
make_genresList(data['genres'][2])
```

```python
data['genres_list'] = data['genres'].map(lambda x : make_genresList(x))
data['genres_list']
```

```python
data.columns
```

```python
# grep actor list and return nth actor
def get_actor(x, n):
    n = n-1 # converted to list index
    for index, i in enumerate(x):
        if index == n:
#             print(i)
            return i['name']

    return np.NaN
```

```python
data['cast'][0]
```

```python
get_actor(data['cast'][0], 1)
```

```python
data['actor_1_name'] = data['cast'].map(lambda x: get_actor(x,1))
data['actor_2_name'] = data['cast'].map(lambda x: get_actor(x,2))
data['actor_3_name'] = data['cast'].map(lambda x: get_actor(x,3))
```

```python
data['actor_1_name'][0], data['actor_2_name'][0], data['actor_3_name'][0]
```

```python
data['crew'][0]
```

```python
def get_directors(x):
    dt = []
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (' '.join(dt))
```

```python
get_directors(data['crew'][0])
```

```python
data['director_name'] = data['crew'].map(lambda x: get_directors(x))
data['director_name'].head()
```

```python
data.columns
```

```python
movie = data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres_list','title']]
movie.head()
```

```python
# checking nan values
movie.isna().sum()
```

```python
# droping nan data points
movie = movie.dropna(how='any')
```

```python
movie = movie.rename(columns={'genres_list':'genres'})
movie = movie.rename(columns={'title':'movie_title'})
movie.head()
```

```python
movie['movie_title'] = movie['movie_title'].str.lower()
```

```python
movie['comb'] = movie['actor_1_name'] + ' ' + movie['actor_2_name'] + ' ' + movie['actor_3_name'] + ' '+ movie['director_name'] + ' ' + movie['genres']

movie.head()
```

```python
movie.drop_duplicates(subset ="movie_title", keep = 'last', inplace = True)
```

```python
movie.to_csv('transformed_kaggle_data.csv',index=False)
```
